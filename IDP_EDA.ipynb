## IDP's Support

The problem of protecting and assisting IDPs is not a new issue. In international law it is the responsibility of the government concerned to provide assistance and protection for the IDPs in their country. 
However, as many of the displaced are a result of civil conflict and violence or where the authority of the central state is in doubt, there is no local authority willing to provide assistance and protection.(Goodwin-Gill, Guy S. (1996)) 
It has been estimated that some 5 million IDPs in 11 countries are "without any significant humanitarian assistance from their governments." (IDMC (April 2006))

```python
data.head(2)
```

```python
def supp(df):
    x= df.value_counts(normalize=True)
    y = pd.DataFrame(x)
    return y
```

## Analysis of the people of IDPs

#### Have the idps being displaced before?

```python
data['Have IDPs been displaced previously'].value_counts().plot.pie(colors = ['lightblue', 'lightgreen'])

plt.title('Have IDPs been displaced previously', fontweight = 30, fontsize = 20)
plt.xlabel('status')
plt.ylabel('count')
plt.show()
```

#### Why were the people displaced?

```python
plt.rcParams['figure.figsize'] = (15, 9)
x = pd.crosstab(data['Reason for displacement of Majority'], data['State'])
x.div(x.sum(1).astype(float), axis = 0).plot(kind = 'bar', stacked = False)
plt.title('Reason for displacement by state', fontweight = 30, fontsize = 20)
plt.show()
```

### Where do the people plan to return to when the reason for displacement has been settled?

In so-called "post-conflict" situations, there has traditionally been an emphasis in the international community to seek to return to the pre-war status-quo (Sara Pantuliano (2009)). However, opinions are gradually changing, because violent conflict destroys political, economic and social structures and new structures develop as a result, quite often irreversibly. Furthermore, returning to the pre-war status-quo may actually be undesirable if pre-war structures led to the conflict in the first place, or prevented its early resolution. IDPs' and refugees' right of return can represent one of the most complex aspects of this issue. Sara Pantuliano (2009)

```python
plt.rcParams['figure.figsize'] = (15, 9)
x = pd.crosstab(data['State'], data['Intended Area of Return'])
x.div(x.sum(1).astype(float), axis = 0).plot(kind = 'bar', stacked = 'True')
plt.title('Intended Area of Return by State', fontweight = 30, fontsize = 20)
plt.show()
```
